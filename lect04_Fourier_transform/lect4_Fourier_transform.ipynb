1D Fourier transformation

```python
plt.close('all') #Close all plot
#Initialize parameter
w = np.pi/2
a=0
signal = []
scale = 0.0001
```

```python
#Give range of time from -4pi to 4pi
t= np.arange(-4*np.pi, 4*np.pi, scale)

#Computer function f
for i in np.nditer(t):
  a = a+1
  if (i>= -w/2) and (i<=w/2):
    y=1
  else:
    y=0
  signal.append(y)

x = np.array(signal)

#Theorical
f_theory = 1/t
y_theory = (w/(np.pi*w*f_theory))*np.sin(np.pi*w*f_theory)

#Simulation
dt = t[1]-t[0]
y_simulation = np.fft.fft(x)
f_simulation = np.fft.fftfreq(x.size, d=dt)

```

```python
#Three plots
fig = plt.figure(1, figsize=(10,8),frameon=False)
ax1 = fig.add_subplot(131)
ax1.plot(t,x)
ax1.set_xlabel('t',fontsize=18)
ax1.set_ylabel('x(t)', fontsize=18)
ax1.set_title('Original Signal', fontsize=18)
ax1.set_xlim(-10,10) #adjust the max leaving min unchanged

ax2 = fig.add_subplot(132)
ax2.plot(f_theory, np.abs(y_theory),color='g')
ax2.set_xlabel('freq',fontsize=18)
ax2.set_ylabel('Theory_Absolute Y(f)', fontsize=18)
ax2.set_xlim(-5,5)
ax2.set_title('Theory', fontsize=24) #adjust the max leaving min unchanged


ax3 = fig.add_subplot(133)
ax3.plot(f_simulation, np.abs(np.real(y_simulation)*scale),color='r')
ax3.set_xlabel('freq',fontsize=18)
ax3.set_ylabel('Simulation_Absolute Y(f)', fontsize=18)
ax3.set_xlim(-5,5)
ax3.set_title('simulation', fontsize=24) #adjust the max leaving min unchanged
```

Fourier transform pair another example

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
from matplotlib import projections #Close all plot
#Initialize paramrter
plt.close('all')

resx = 120
resy = 120
cx = resx/2
cy = resx/2

#Compute function f
fxy =  [[0 for x in range(resx)] for y in range(resy)]

for i in range(resx):
  for k in range(resy):
    if (k>=(cx-10)) and (k<=(cx+10)) and (i>=(cy-5)) and (i<=(cy+5)):
      fxy[i][k]=1
    else:
      fxy[i][k]=0

#Use fftshift to shift low frequency in the center
FS = np.fft.fftn(fxy)
fig = plt.figure(1, figsize=(10,8),frameon=False)
ax1 = fig.add_subplot(131)
ax1.imshow(fxy,cmap='gray')
ax1.set_title('Original Image', fontsize = 24)
#Turn off tick lables
ax1.set_yticklabels([])
ax1.set_xticklabels([])

ax2 = fig.add_subplot(132)
ax2.imshow((np.abs(np.fft.fftshift(FS))))
ax2.set_title('FT',fontsize=24)

ax3 = fig.add_subplot(133,projection='3d')
X = np.arange(0, 120, 1)
Y = np.arange(0, 120, 1)
X,Y = np.meshgrid(X,Y)
surf = ax3.plot_surface(X,Y,(np.abs(np.fft.fftshift(FS))),cmap='seismic')

plt.show()
```

processing step

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
```

```python
from matplotlib.image import imread
#Read image
im = cv2.imread('brain.png',0)
#Add zeros padding with (-1)^x+y
nim = np.zeros((2*im.shape[0], 2*im.shape[1]),dtype = im.dtype)
bnim = np.zeros((2*im.shape[0], 2*im.shape[1]),dtype = im.dtype)
for i in range(im.shape[0]):
  for k in range(im.shape[1]):
    nim[i][k] = im[i][k]
    bnim[i][k] = ((-1)**(i+k)*im[i][k])

#Apply fft
FS = np.fft.fft2(bnim)
#Create a filter to keep low frequency, example a bish image in frequency domain
filt = np.zeros((2*im.shape[0],2*im.shape[1]),dtype=im.dtype)
center = (im.shape[1], im.shape[0])
radius = 100
cv2.circle(filt, center, radius, (255,255,255),-1)
#Apply fft
G = filt*FS
gt = np.real(np.fft.ifftn(G))
#Multiple with (-1)**(i+k) back
for i in range(gt.shape[0]):
  for k in range(gt.shape[1]):
    gt[i][k] = ((-1)**(i+k))*gt[i][k]

final = np.zeros((im.shape[0],im.shape[1]),dtype = gt.dtype) 
for i in range(final.shape[0]):
  for k in range(final.shape[1]):
    final[i][k] = gt[i][k]

plt.subplot(241)
plt.imshow(im,'gray')
plt.title('Oritginal Image', fontsize=10)
plt.axis("off")

plt.subplot(242)
plt.imshow(im,'gray')
plt.title('Zeros Padding Image', fontsize=10)
plt.axis("off")

plt.subplot(243)
plt.imshow(nim,'gray')
plt.title('(-1)^(x+y)', fontsize=10)
plt.axis("off")

plt.subplot(244)
plt.imshow(bnim,'gray')
plt.title('FT', fontsize=10)
plt.axis("off")

plt.subplot(245)
plt.imshow(filt,'gray')
plt.title('Filter', fontsize=10)
plt.axis("off")

plt.subplot(246)
plt.imshow(20*np.log(np.abs(G)+1),'gray')
plt.title('After Filtering', fontsize=10)
plt.axis("off")

plt.subplot(247)
plt.imshow(final,'gray')
plt.title('Final Result', fontsize=10)
plt.axis("off")

plt.show()
```
